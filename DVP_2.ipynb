```python
from matplotlib import pyplot as plt
```

```python
dev_ages  =[21, 22, 23, 24]
dev_sal = [45000, 50000, 60000, 75000]
dev_python = [50000, 60000, 80000, 100000]

plt.title("Average sales in $")
plt.xlabel('Age')
plt.ylabel('Salary')

plt.plot(dev_ages, dev_sal, color = 'green', linestyle = '-', label = 'Developer Salary')
plt.plot(dev_ages, dev_python, color = 'red', linestyle = '--', label = 'Python Developer Salary')

plt.legend()
```
